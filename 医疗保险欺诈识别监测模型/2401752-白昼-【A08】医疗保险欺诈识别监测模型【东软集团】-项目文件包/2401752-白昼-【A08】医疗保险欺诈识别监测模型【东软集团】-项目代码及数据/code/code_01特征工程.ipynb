### 评估数据干净度

```python
ori_data.info()
```

- 第71个特征：住院诊断LENTH_MAX 存在缺失值
- 第0个特征：个人编码全部一样，可以删除
- 第1~19个特征：天数、次数、金额等，确实为数值变量，且分布较均匀，该部分特征初步判断无较大问题
- 第20个特征：医院编码_NN 应为分类变量，这里为int64数值类型

- 出院诊断病种名称、民政救助、城乡优抚、是否挂号为分类变量，应进行独热编码

# 清洗数据

## 删除无意义特征

```python
clean_data = ori_data.drop(columns=['个人编码','顺序号_NN','交易时间DD_NN','交易时间YYYY_NN','交易时间YYYYMM_NN'])
print(clean_data.shape)
print(ori_data.shape)
```

## 删除重复数据

```python
clean_data.drop_duplicates(inplace=True)
clean_data.reset_index(drop=True, inplace=True)
clean_data.shape
```

## 缺失值处理

```python
clean_data['出院诊断LENTH_MAX'].describe()
```

```python
from sklearn.impute import SimpleImputer
lenth_max = clean_data.loc[:, "出院诊断LENTH_MAX"].values.reshape(-1, 1)
clean_data['出院诊断LENTH_MAX'].values
imp_median = SimpleImputer(strategy='median')
imp_median = imp_median.fit_transform(lenth_max)
clean_data.loc[:, "出院诊断LENTH_MAX"] = imp_median
```

## 独热编码

```python
from sklearn.preprocessing import OneHotEncoder
X_onehot = clean_data.loc[:, ["医院编码_NN", "出院诊断病种名称_NN"]]
enc = OneHotEncoder(categories='auto').fit(X_onehot)
result_onehot = enc.transform(X_onehot).toarray()
feature_name = enc.get_feature_names_out()
data_onehot = pd.DataFrame(result_onehot, columns=feature_name)
newdata = pd.concat([clean_data, data_onehot],axis=1)
newdata.drop(columns=["医院编码_NN", "出院诊断病种名称_NN"], inplace=True)
```

## 归一化

```python
from sklearn.preprocessing import MinMaxScaler
data_minmax = pd.DataFrame(MinMaxScaler().fit_transform(newdata),columns = newdata.columns )
# 将要移动的列使用pop方法取出，然后放到最后
res = data_minmax.pop('RES')

# 重新创建DataFrame，并将取出的列作为最后一列
data_minmax = pd.concat(
    [data_minmax, pd.DataFrame(res, columns=["RES"])], axis=1)
data_minmax.head()
```

## 低方差滤波

```python
var = data_minmax.var()
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()  
data_var0 = selector.fit_transform(data_minmax)  # 获取删除不合格特征之后的新特征矩阵
minmax_columns = data_minmax.columns
data_var0_df = pd.DataFrame(
    data_var0, columns=minmax_columns[selector.get_support(indices=True)])
```

通过低方差滤波过滤掉了方差为0的特征 住院天数

## 互信息法

```python
from sklearn.feature_selection import mutual_info_classif as MIC
X_0 = data_var0_df.iloc[:,:-1]
y = data_var0_df.iloc[:,-1]
result = MIC(X_0, y,random_state=0)
k = result.shape[0] - sum(result <= 0)
data_var0_columns = data_var0_df.iloc[:, :-1].columns
data_mic_df = pd.DataFrame(
    X_0.iloc[:, result > 0], columns=data_var0_columns[result > 0]
)
X_mic = X_0.iloc[:, result > 0]
data_mic_df["RES"] = y
# data_mic_df.to_excel('data_预处理1.xlsx', index=False)
```
