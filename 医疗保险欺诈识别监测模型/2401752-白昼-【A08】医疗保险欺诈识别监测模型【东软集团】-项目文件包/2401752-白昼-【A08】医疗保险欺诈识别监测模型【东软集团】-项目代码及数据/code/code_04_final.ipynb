```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import classification_report  # 衡量分类模型准确度，生成分类模型性能报告
import pandas as pd

data_test = pd.read_excel('../data/data_test.xlsx')
data_train = pd.read_excel('../data/data_train_smote.xlsx')
rfc = RFC(n_estimators=143
          , max_depth=29
            ,min_samples_leaf=1
            ,min_samples_split=2
          , random_state=100
          )

rfc.fit(data_train.iloc[:,:-1],data_train.iloc[:,-1])
X_test = data_test.iloc[:, :-1]
y_test = data_test.iloc[:, -1]
y_pre = rfc.predict(X_test)

print('Classification Report:\n', classification_report(y_test, y_pre, digits=4))
```
