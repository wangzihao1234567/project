# 划分训练集测试集

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_excel('../data/data_handle.xlsx')
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=100, test_size=0.2
    # , shuffle=False
)
df_train = pd.concat([X_train, y_train], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)
df_train.to_excel('data_train.xlsx', index=False) # 训练集
df_test.to_excel('data_test.xlsx', index=False) # 测试集
```

## smote过采样

```python
from imblearn.over_sampling import SMOTE
# 未过采样前训练集

X_ori = df_train.iloc[:, :-1]
y_ori = df_train.iloc[:, -1]
smote = SMOTE(
    random_state=100, k_neighbors=16, sampling_strategy=0.18
)
x_smote, y_smote = smote.fit_resample(X_ori, y_ori)
data = pd.concat([x_smote, y_smote], axis=1)
data.to_excel('data_train_smote.xlsx', index=False)
```
